# AirQuality Data Preparation

```python
import pandas as pd
import matplotlib
import seaborn as sns
import warnings
import os

from matplotlib import pyplot as plt

pd.set_option('display.max_columns', None)

print('Pandas version:', pd.__version__)
print('Matplotlib version:', matplotlib.__version__)
print('Seaborn version:', sns.__version__)

from datetime import date
from dateutil.relativedelta import relativedelta

pd.set_option('display.max_columns', None)
warnings.filterwarnings('ignore')
```

## Loading the data

```python
data_dir = "./data/location/"

dfs = []

for file in os.listdir(data_dir):
    file_path = os.path.join(data_dir, file)

    if not file.endswith('.csv'):
        continue

    df = pd.read_csv(file_path, sep=',')

    sensor_id = df.columns[1].split("-")[0]
    df['Sensor_ID'] = sensor_id

    new_columns = [col.split("-")[-1] if "-" in col else col for col in df.columns]
    df.columns = new_columns

    dfs.append(df)

combined_df = pd.concat(dfs, ignore_index=True)
combined_df.sample(15)
```

```python
combined_df = combined_df.rename(columns={'Tijd': 'Date', 'Lon': 'Longitude', 'Lat': 'Latitude'})

combined_df['Date'] = pd.to_datetime(combined_df['Date'])
combined_df['Date'] = combined_df['Date'].dt.date

combined_df
```

```python
output_file = "data/combined_data.csv"

combined_df.to_csv(output_file, index=False)
```
